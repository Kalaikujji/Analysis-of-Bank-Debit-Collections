```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Walmart_sales.csv")
```

```python
wal_sales = data.copy()
wal_sales.head()
```

```python
wal_sales.shape
```

```python
wal_sales.dtypes
```

```python
from pickle import TRUE
# Top 10 store with most sales
pd.set_option('display.float_format', lambda x: '%.3f' % x)

top_10 = wal_sales.groupby('Store') \
                .agg({'Weekly_Sales' : 'sum'}) \
                .sort_values( by = 'Weekly_Sales', ascending = True) \
                .head(10) \
                .rename(columns = {'Weekly_Sales' : 'Total_Sales'})
```

```python
# creating bar graph for top 10 most selling stores
top_10.plot(kind = 'bar', y = 'Total_Sales', legend = True)
plt.xlabel('Store_no')
plt.ylabel('Total Sales')
plt.title('Top 10 Stores by Total Sales')

plt.show()
```

```python
# Is there any relation between temperature and weeklysales?
#let's find out
wal_sales["Temperature"].corr(wal_sales['Weekly_Sales'])
# There is almost no correlation between temp and weekly sales.
```

```python
wal_sales["Fuel_Price"].corr(wal_sales["Weekly_Sales"])
```

```python
# Most stores tend to have more sales on holiday. Is that's the case for walmart? Let's find out.

#average sales on holiday for each store.
avg_hol_sales = wal_sales.loc[ wal_sales['Holiday_Flag'] == 1] \
                        .groupby('Store') \
                        .agg({'Weekly_Sales' : 'mean'}) \
                        .rename(columns = {'Weekly_Sales' : 'Avg_sal_holiday'}) # average sales on holiday.
avg_hol_sales.head(10)
```

```python
avg_sal_workday = wal_sales.loc[ wal_sales['Holiday_Flag'] == 0] \
                            .groupby('Store') \
                            .agg({'Weekly_Sales' : 'median'}) \
                            .rename(columns = {'Weekly_Sales' : 'Avg_sal_workday'})

avg_sal_workday.head(10)
```

```python
# from the above two outputs it seems on holidays walmart have more sales on average.
#let's inspect that visually.
sales = pd.concat([avg_hol_sales, avg_sal_workday ], axis = 1)
sales.head(20) \
    .plot.bar()

# As we can observe visually for the first 10 stores Average sales on holiday is higher.
# They can focus investing more on their holiday advertisment campaign.
```
